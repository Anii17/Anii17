```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
```

```python
path='/content/drive/MyDrive/Iris classification /iris.csv'
dataset = pd.read_csv(path)
```

```python
dataset.head()
```

```python
%matplotlib inline
img=mpimg.imread('/content/drive/MyDrive/Iris classification /iris_types.jpg')
plt.figure(figsize=(20,40))
plt.axis('off')
plt.imshow(img)
```

```python
X = dataset.iloc[:,:4].values
y = dataset['species'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 82)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

*Linear kernel*

```python
from sklearn.svm import SVC
svcclassifier = SVC(kernel = 'linear', random_state = 0)
svcclassifier.fit(X_train, y_train)
```

```python
y_pred = svcclassifier.predict(X_test)
print(y_pred)
```

```python
#lets see the actual and predicted value side by side
y_compare = np.vstack((y_test,y_pred)).T
#actual value on the left side and predicted value on the right hand side
#printing the top 5 values
y_compare[:5,:]
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
#finding accuracy from the confusion matrix.
a = cm.shape
corrPred = 0
falsePred = 0

for row in range(a[0]):
    for c in range(a[1]):
        if row == c:
            corrPred +=cm[row,c]
        else:
            falsePred += cm[row,c]
print('Correct predictions: ', corrPred)
print('False predictions', falsePred)
kernelLinearAccuracy = corrPred/(cm.sum())
print ('Accuracy of the SVC Clasification is: ', corrPred/(cm.sum()))       
```

*POLYNOMIAL KERNEL*

```python
from sklearn.svm import SVC
svcclassifier = SVC(kernel = 'poly', random_state = 0)
svcclassifier.fit(X_train, y_train)
```

```python
y_pred = svcclassifier.predict(X_test)
print(y_pred)
```

```python
#lets see the actual and predicted value side by side
y_compare = np.vstack((y_test,y_pred)).T
#actual value on the left side and predicted value on the right hand side
#printing the top 5 values
y_compare[:5,:]
```

*Radial Basis Function*

# ***RANDOM FOREST CLASSIFIER - IRIS DATASET***

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
print(iris.target_names)

print(iris.feature_names)
```

```python
print(iris.data[0:5])
print(iris.target)
```

```python
import pandas as pd
data=pd.DataFrame({
    'sepal length':iris.data[:,0],
    'sepal width':iris.data[:,1],
    'petal length':iris.data[:,2],
    'petal width':iris.data[:,3],
    'species':iris.target
})
data.head()
```

```python
from sklearn.model_selection import train_test_split

x=data[['sepal length', 'sepal width', 'petal length', 'petal width']]
y=data['species']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf=RandomForestClassifier(n_estimators=100)


clf.fit(x_train, y_train)

y_pred=clf.predict(x_test)
```

```python
from sklearn import metrics

print("accuracy:",metrics.accuracy_score(y_test, y_pred))
```
